```python
# Importing the libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Uploading the dataset
dataset = pd.read_csv("/content/creditcard.csv")
```

```python
# Printing the head of the dataset
dataset.head()
```

```python
# Printing the tail of the dataset
dataset.tail()
```

```python
# Getting the shape of the dataset
dataset.shape
```

```python
# Describing the dataset
dataset.describe()
```

```python
# Getting the info of the dataset
dataset.info()
```

```python
# Checking for missing values
dataset.isnull().sum()
```

```python
# Counting the Class of the dataset
dataset['Class'].value_counts()
```

1: Fraud

2: Legit


```python
# Comparing the values of both classes
dataset.groupby("Class").mean()
```

```python
# Creating a new dataset
legit = dataset[dataset.Class == 0]
fraud = dataset[dataset.Class == 1]
```

```python
# Creating a legit sampple
legit_sample = legit.sample(n=492)
```

```python
# Concat the legit sample and the fraud class
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
# Counting the value of the new dataset
new_dataset['Class'].value_counts()
```

```python
# Getting the mean of the new dataset
new_dataset.groupby("Class").mean()
```

```python
# Separating the data and class
X = new_dataset.drop(columns="Class", axis=1)
Y = new_dataset["Class"]
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting train and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# Printing the shape of the train and test data
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Training Logistic Regression model
model = LogisticRegression()
```

```python
# Fitting the model
model.fit(X_train, Y_train)
```

```python
# Evaluating the model (train)
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print(f"Train accuracy: {training_data_accuracy}")
```

```python
# Evaluating the model (test)
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print(f"Test accuracy: {test_data_accuracy}")
```

```python
# Comparing the predicted values and actual values with two plots
plt.plot(Y_test.values, label='Actual Values', color='blue')
plt.plot(X_test_prediction, label='Predicted Values', color='red')
plt.xlabel('Data Points')
plt.ylabel('Gold Price')
plt.title('Actual vs. Predicted Gold Prices')
plt.legend()
plt.show()
```
